```python
import os
import shutil
```

```python
train_folder = 'training_data'
test_folder = 'test_data'
exception_folders = ['character', 'nac']
```

```python
tester = input()
```

```python
tester
```

```python
old_tester = os.listdir(os.path.join(test_folder, 'a'))
```

```python
old_tester[0]
```

```python
gestures = os.listdir(train_folder)
```

```python
for g in gestures:
    if g not in exception_folders:
        target = os.path.join(train_folder, g, old_tester[0])
        original = os.path.join(test_folder, g, old_tester[0])
        shutil.move(original, target)
```

```python
for g in gestures:
    if g not in exception_folders:
        original = os.path.join(train_folder, g, tester)
        target = os.path.join(test_folder, g, tester)
        shutil.move(original, target)
```
