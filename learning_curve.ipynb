```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train_history.csv')
```

```python
data.head()
```

```python
l_curve = data[['loss', 'accuracy', 'val_loss', 'val_accuracy']]
l_curve.head()
```

```python
l_curve.plot(figsize=(8, 5))
plt.grid(True)
plt.gca().set_ylim(0, 1) # set the vertical range to [0-1]
plt.xlabel('Epochs')
plt.savefig('learning_curve.png')
plt.show()
```

```python
plt.plot(data.loss)
plt.plot(data.val_loss)
```

```python
plt.plot(data.accuracy)
plt.plot(data.val_accuracy)
```
