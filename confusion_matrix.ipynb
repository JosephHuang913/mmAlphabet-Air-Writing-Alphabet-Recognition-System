```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
conf_matrix = np.load('confusion_matrix.npy')
```

```python
conf_matrix
```

```python
conf_matrix_f = np.zeros_like(conf_matrix, dtype=float)
```

```python
for i in range(conf_matrix.shape[0]):
    #print(conf_matrix[i, :] / float(conf_matrix[i].sum()))
    conf_matrix_f[i, :] = conf_matrix[i, :] / float(conf_matrix[i].sum())
```

```python
conf_matrix_f
conf_matrix_df = pd.DataFrame(conf_matrix_f)
conf_matrix_df
```

```python
conf_matrix_df.to_csv('confusion_matrix.csv')
```

```python
img = plt.imshow(conf_matrix_f)
plt.colorbar(img)
plt.savefig('confusion_matrix.png')
```
